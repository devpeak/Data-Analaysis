***Function of Sepals***

Protects the reproductive organs within the flower. In some plants with no petals, sepals function as petals. Safeguard the bud and flowers from harsh environmental conditions. Sepals protect both flowers and fruits by producing chemicals that would ward off predators.

***Function of Petals***

Petals are coloured, modified leaves that surround the reproductive parts of flowers. They are brightly coloured to attract insects for pollinators. Therefore, their function is to help in pollination while also protecting the inner set of essential whorls

***Seaborn is a Python data visualization library based on matplotlib. It provides a high-level interface for drawing attractive and informative statistical graphics.***

```python
#2D Scatter Plot with color-coding for each flower type/class.
#Here 'sns' corresponds to *searborn*

sns.set_style('whitegrid');
sns.FacetGrid(iris, hue = 'species', height = 4) \
    .map(plt.scatter, 'sepal_length','sepal_width') \
    .add_legend();
plt.show();
```

***sns.set_style('whitegrid');***
Parameters

dict, or one of {darkgrid, whitegrid, dark, white, ticks}

A dictionary of parameters or the name of a preconfigured style.

https://seaborn.pydata.org/generated/seaborn.FacetGrid.html

Variables that define subsets of the data, which will be drawn on separate facets in the grid.

Height (in inches) of each facet
